# Imports

```python
import os
from PIL import Image
```

# Pegando os arquivos

```python
path = '../data/dogs-vs-cats/train/'
files = os.listdir(path)
```

# Separando

```python
dogs, cats = [], []
for f in files:
    if 'cat' in f:
        cats.append(f)
    elif 'dog' in f:
        dogs.append(f)
```

# Escolhendo

```python
dogs_chosen = dogs[:1200]
cats_chosen = cats[:1200]
```

# Salvando escolhidos em outra pasta

```python
for c, d in zip(cats_chosen, dogs_chosen):
    Image.open(f'{path}{c}').save(f'../data/cats/{c}')
    Image.open(f'{path}{d}').save(f'../data/dogs/{d}')
```

# Checando

```python
Image.open(f'../data/cats/{cats_chosen[0]}')
```

```python
Image.open(f'../data/dogs/{dogs_chosen[0]}')
```
