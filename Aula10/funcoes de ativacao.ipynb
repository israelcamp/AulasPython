# Aula 10 - Funcoes de ativacao

Em redes neurais, uma das principais escolhas a se fazer e em relacao a quais funcoes de ativacao iremos utilizar. Logo, vamos mostrar aqui o comportamento de duas das funcoes mais utilizadas, as funcoes sigmoidais e a funcoes retificadoras.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-10, 10, 1000)
```

### Funcoes sigmoidas

```python
def sigmoid(x):
    return 1. / (1. + np.exp(-x))

def tanh(x):
    return np.tanh(x)
```

```python
plt.figure(figsize=(10,10))
plt.plot(x, sigmoid(x))
plt.plot(x, tanh(x))
plt.legend(['Sigmoid', 'Tanh'])
plt.grid()
```

Funcoes sigmoidais sao mais utilizadas na camada de saida.

## Funcoes retificadoras

```python
def relu(x):
    return x * (x > 0)

def leaky_relu(x, alpha = 0.05):
    y = x.copy()
    
    y[y < 0] = y[y < 0] * alpha
    return y

def relu6(x):
    z = x.copy()
    z = relu(z)
    z[z > 6] = 6
    return z
```

```python
f, ax = plt.subplots(2, sharex=True, figsize=(10,10))
ax[0].plot(x, relu(x))
ax[0].legend(['ReLU'])
ax[0].grid()
ax[1].plot(x, leaky_relu(x))
ax[1].plot(x, relu6(x))
plt.legend(['LeakyReLU', 'ReLU6'])
plt.grid()
```

Funcoes retificadoras sao importantes em redes convolucionais e sao bastante utilizadas em camadas escondidas
